```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('income_evaluation.csv')
df
```

```python
df.columns
```

```python
df.columns = [col if i == 0 else col.replace(' ', '') for i, col in enumerate(df.columns)]
```

```python
df.columns
```

```python
y = df['income']
x = df.drop('income', axis = 1)
```

```python
y
```

```python
#y_final = y.apply(lambda x: 1 if x=='>50K' else 0)

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

y_final = label_encoder.fit_transform(y)
```

```python
y_final
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

numerical = ['age', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']
x[numerical] = scaler.fit_transform(x[numerical])
x
```

```python
x_final = pd.get_dummies(x)
```

```python
x_final
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_final, y_final, test_size = 0.25, random_state = 1)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
cls=GradientBoostingClassifier(n_estimators=500, learning_rate=0.05, random_state=0)
```

```python
cls.fit(x_train,y_train)
```

```python
y_pred = cls.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
accuracy = accuracy_score(y_test,y_pred)
```

```python
accuracy
```

```python
report=classification_report(y_test,y_pred)
print(report)
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.model_selection import GridSearchCV
grid = {'learning_rate':[0.01,0.05,0.1], 'n_estimators':np.arange(100,500,100),}
```

```python
gb_cv = GridSearchCV(cls, grid, cv = 4)
gb_cv.fit(x_train,y_train)
```

```python
print("Best Parameters:",gb_cv.best_params_)
print("Train Score:",gb_cv.best_score_)
print("Test Score:",gb_cv.score(x_test,y_test))
```

```python
grid = {'learning_rate':[0.01,0.05,0.1], 'n_estimators':np.arange(100,500,100),'max_depth':[2,3,4,5,6,7] }
```

```python
gb_cv = GridSearchCV(cls, grid, cv = 4)
gb_cv.fit(x_train,y_train)
```
